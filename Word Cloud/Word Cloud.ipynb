```python
words = txt.split(' ')
print("Total words: ", len(words))
print("Vocab Size: ",len(list(set(words))))

vocab = list(set(words))
```

```python
for word in vocab:
    if(word == 'KNOWN'.lower()):
        print(word)
```

```python
def counter(word):
    
    c = 0
    for i in words:
        if (i == word):
            c += 1
            
    return c
```

```python
word_counter = {}

for i in vocab:
    word_counter[i] = counter(i)
```

```python
word_2_int = {}
int_2_word = {}
var = 0

for i in vocab:
    
    word_2_int[i] = var
    int_2_word[var] = i
    
    var += 1
```
